```python
recipe_df = pd.read_csv("레시피+과정정보.csv")
```

```python
import pandas as pd

# 레시피 데이터프레임
final_df = pd.DataFrame(columns = ['title', 'ingredients', 'recipe'])
```

```python
for i in range(len(title_df)):
    ID = title_df['RECIPE_ID'][i]
    title = title_df['RECIPE_NM_KO'][i]
    
    ingredient = ingredient_df[ingredient_df['RECIPE_ID'] == ID]['IRDNT_NM'].tolist()
    
    recipe = recipe_df[recipe_df['RECIPE_ID'] == ID]['COOKING_DC'].tolist()
    recipe = "".join(recipe)
    
    final_df.loc[i] = [title, ingredient, recipe]
```

```python
final_df.tail()
```

```python
final_df.to_csv('recipe_data_2.csv', index=False, encoding='utf-8-sig')
```
