```python
import time
import requests
from bs4 import BeautifulSoup

from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

url = "https://wtable.co.kr/recipes" # 우리의 식탁

driver = Chrome()
driver.set_window_size(1920,1080)
driver.get(url)
```

```python
import pandas as pd

# 레시피 데이터프레임
recipe_df = pd.DataFrame(columns = ['title', 'ingredients', 'recipe'])
```

```python
menu_pics = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,'p.RecipeItemstyle__Title-sc-1gt7vz8-3.hpYiJK')))
```

```python
print(len(menu_pics))
```

```python
str = '\n'
i = 0

while i != 1743:
    try:
        menu_pics[i].click() # 레시피 사진 클릭
        time.sleep(4)

        # 요리 명
        title = driver.find_elements_by_css_selector('h2.RecipeDetailstyle__Title-q7sykd-4.kIVrZW')[0].text

        # 재료 리스트
        ingredients = []
        ingredients_tag = driver.find_elements_by_css_selector('div.Text__Description02-sc-1qy6bx2-0.fCbbYE')
        for ingredient in ingredients_tag:
            index = ingredient.text.find(str)
            if index != -1:
                ingredients.append(ingredient.text[:index])
            else:
                ingredients.append(ingredient.text)

        # 레시피
        recipe = ''
        recipes = driver.find_elements_by_css_selector('p.Text__Pre01-sc-1qy6bx2-2.hJDmSV')
        for temp in recipes:
            recipe += temp.text

        # 데이터프레임에 추가
        recipe_df.loc[i] = [title, ingredients, recipe]
        print(i)
        time.sleep(1)
        
        # 뒤로가기
        driver.back()
        time.sleep(4)
        i+= 1
    except:
        print('pass',title)
        pass
    
    if i != 0 and i%4 == 0:
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        time.sleep(4)
```
