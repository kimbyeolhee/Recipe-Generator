```python
import time
import requests
from bs4 import BeautifulSoup

from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

url = "https://2bob.co.kr/recipe.php?id=list&eTheme=1" # 이밥차

driver = Chrome()
driver.set_window_size(1080,1080)
driver.get(url)
```

```python
import pandas as pd

# 레시피 데이터프레임
recipe_df = pd.DataFrame(columns = ['title', 'ingredients', 'recipe'])
final_df = pd.DataFrame(columns = ['title', 'ingredients', 'recipe'])
```

```python
while True:
    time.sleep(3)
    menu_pics = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,'p.b_title')))
    recipe_df = pd.DataFrame(columns = ['title', 'ingredients', 'recipe'])
    
    for i in range(len(menu_pics)):
        try:
            menu_pics = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,'p.b_title')))
            menu_pics[i].click()
            print(i)
            time.sleep(4)

            # 요리명
            title = driver.find_elements_by_css_selector('h2.prod_title')[0].text

            # 재료 리스트
            ingredients = []
            str = '('
            preprocessed_ingredients = []

            ingredients_tag = driver.find_elements_by_css_selector('p.mate_list')

            for ingredient in ingredients_tag:
                ingredients += ingredient.text.split(',')

            for ingredient in ingredients:
                index = ingredient.find(str)
                if index == -1:
                    preprocessed_ingredients.append(ingredient.strip())
                else:
                    preprocessed_ingredients.append(ingredient[:index].strip())

            # 레시피
            recipe = ''
            recipes_tag = driver.find_elements_by_css_selector('div.text_exp')

            for recipe_text in recipes_tag:
                index = recipe_text.text.find(str)
                if index == -1:
                    recipe += recipe_text.text.replace('\n', ' ')
                else:
                    recipe += recipe_text.text[:index].replace('\n', ' ')

            # 데이터프레임에 추가
            recipe_df.loc[i] = [title, preprocessed_ingredients, recipe]
            print(i, title)

            # 뒤로가기
            driver.back()
            time.sleep(3)
        except:
            pass
        
    if len(menu_pics) != 20:
        break
    else:
        # 다음 페이지로 넘어가기
        final_df = pd.concat([final_df, recipe_df], ignore_index=True)
        next_btn = driver.find_elements_by_css_selector('p.paging_btn')
        if len(next_btn) == 4:
            next_btn[2].click()
        elif len(next_btn) == 3:
            next_btn[2].click()
        elif len(next_btn) == 2:
            next_btn[1].click()
        time.sleep(4)
```
