```python
title = "라자냐"
ingredients = ["치즈","스파게티소스","케챱","계란"]
```

`<unused0>` , `<unused1>` : 요리이름의 시작과 끝 <br/>
`<unused2>` , `<unused3>` : 재료이름의 시작과 끝 <br/>

```python
def input_format(title, ingredients):
  """ transform user's input to fit model's input form
      Args:
        title: 요리 이름
        ingredients: 재료들 => list
      
      return:
        <unused0>title<unused1><unused2>재료1$재료2$재료3<unused3>
  """

  temp1 = '<unused0>' + title + '<unused1>'
  temp2 = '<unused2>'
  for ingredient in ingredients:
    temp2 += ingredient + '$'
  temp2 += '<unused3>'
  return temp1 + temp2
```

```python
input_format("김치볶음밥", ["치즈", "김치", "밥"])
```
