# train dataset.txt 형식으로 변환

```python
len(temp)
```

```python
def df_to_txt(filename, df):
  f = open(f'{filename}.txt', 'w')

  for i in range(len(df)):
    to_write = ''

    title = df['title'][i]
    to_write += '<unused0>'+title+'<unused1>'
    ingredients = df['ingredients'][i].replace('[','').replace(']','').replace("'","").replace(' ','')
    ingredients = ingredients.split(',')
    to_write += '<unused2>'+'$'.join(ingredients)+'$<unused3>'
    recipe = df['recipe'][i]
    to_write += '<unused4>'+ str(recipe) +'<unused5>'
    to_write += '\n'

    f.writelines(to_write)
    print(i)
  f.close()
```

```python
df_to_txt('crawling_data_2', temp)
```
