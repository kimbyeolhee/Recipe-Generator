# ingredient dictionary

```python
# train dataset

lines = open("/content/drive/MyDrive/11_recipe_generator/data/train_3.txt", encoding="utf-8").readlines()

ingredient_dict = []

for i,line in enumerate(lines):
  start = line.find('<unused2>')
  end = line.find('<unused3>')

  ingredients = line[start+9:end-1].split('$')
  
  for ing in ingredients:
    if ing not in ingredient_dict:
      ingredient_dict.append(ing)

print(ingredient_dict)
```

```python
# valid dataset

lines = open("/content/drive/MyDrive/11_recipe_generator/data/val_preprocessed.txt", encoding="utf-8").readlines()

for i,line in enumerate(lines):
  start = line.find('<unused2>')
  end = line.find('<unused3>')

  ingredients = line[start+9:end-1].split('$')

  for ing in ingredients:
    if ing not in ingredient_dict:
      ingredient_dict.append(ing)
```

```python
# test dataset
lines = open("/content/drive/MyDrive/11_recipe_generator/data/test_for_custom_metric.reg.txt", encoding="utf-8").readlines()

for i,line in enumerate(lines):
  start = line.find('<unused2>')
  end = line.find('<unused3>')

  ingredients = line[start+9:end-1].split('$')

  for ing in ingredients:
    if ing not in ingredient_dict:
      ingredient_dict.append(ing)
```

```python
with open('ingredient_list.txt','w',encoding='UTF-8') as f:
    for ing in ingredient_dict:
        f.write(ing+'\n')
```
